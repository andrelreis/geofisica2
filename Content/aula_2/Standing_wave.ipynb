# A Onda estacionária

As ondas estacionárias são ondas que possuem um mesmo padrão de vibração, que se formam a partir do princípio da superposição de duas ondas idênticas que se propagam em sentidos opostos, mas com a mesma frequência e mesma amplitude. Estas ondas são caracterizadas por pontos fixos que são chamados de nós, bem como pontos de máximo também fixos.

## Breve descrição matemática

Em um movimento hamônico simples, como por exemplo a propagação de onda em uma corda, considerando também uma amplitude fixa, podemos descrever o deslocamento de uma onda como : 

$$ u(t) = A \cos(\omega t + \theta_0) $$
em que,

* A é a amplitude da onda;
* $\omega$ é a frequência angular (em radianos por segundo);
* $\theta_0$ é a fase;
* t é o tempo.
